```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import LSTM, Dense, Flatten,Dropout, Embedding
```

```python
data = pd.read_csv('./movies_genres.csv',sep='\t')
```

```python
data.head()
```

```python
genres = list(data.drop(['title', 'plot'], axis=1).columns.values)
```

```python
x_list = data['plot'].values
```

```python
x_list = list(x_list)
x_list = x_list[0:100]
len(x_list)
x_list2 = x_list[100:120]
```

```python
for i in range(len(x_list)):
    wor = x_list[i].split()
    words = [w for w in wor if w not in stopwords.words('english')]
    string = ' '.join(words)
    x_list[i] = string
```

```python
y_list = np.array(data.drop(['title', 'plot'],axis=1))
```

```python
x_list=np.array(x_list)
len(x_list)
```

```python
y_list = y_list[0:100]
```

```python
tokenizer = Tokenizer(filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n', split=" ")
```

```python
tokenizer.fit_on_texts(x_list)
```

```python
x_train = tokenizer.texts_to_sequences(x_list)
x_test = tokenizer.texts_to_sequences(x_list2)
```

```python
tokenizer.word_index
```

```python
x_train = pad_sequences(x_train,maxlen=500, padding='pre')
```

```python
n_feat = len(tokenizer.word_index)+1
labels=len(y_list[0])
```

```python
model = Sequential()
model.add(Embedding(input_dim=n_feat, output_dim=100, input_length=500))
model.add(Dropout(0.3))
model.add(LSTM(64, return_sequences = True))
model.add(Dropout(0.3))
model.add(LSTM(64))
model.add(Dropout(0.3))
model.add(Dense(labels, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
h = model.fit(x_train,y_list, epochs=5,verbose=1)
```

```python
x_test
```

```python
model.predict(np.array(x_test))
```
